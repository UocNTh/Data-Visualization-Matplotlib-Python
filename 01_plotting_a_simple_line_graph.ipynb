### **Khái niệm chung** 

Một Matplotlib figure có thể được phân loại thành nhiều phần như dưới đây: 

- Figure: Như một cửa sổ chứa tất cả những gì bạn sẽ vẽ trên đó 

- Axes: Thành phần chính của một figure là các axes (những khung nhỏ hơn để vẽ hình trên đó). Một figure có thể chứa một hoặc nhiều axes. Figure chỉ là khung chứa, chính các axes mới thật sự là nơi các hình vẽ được vẽ lên.

- Axis: Chúng là những dòng số giống như các đối tượng và đảm nhận việc tạo các giới hạn biểu đồ 


- Artist: Mọi thứ có thể nhìn thấy trên figure là một artist như Text objects, Line2D objects, collection objects. Hầu hết các Artists được gắn với Axes.

### **Bắt đầu với Pyplot** 


Pyplot là một module của Matlpotlib cung cấp các hàm đơn giản để thêm các thành phần plot như lines, images, text, v.v vào các axes trong figure. 



#### **Plotting a Simple Line Graph**



- Nhập module bằng bí danh plt để hạn chế việc phải nhập pyplot nhiều lần.
- Hàm ```subplots()```: Hoạt động như một trình quản lý trục tự động 
- ```fig, ax = plt.subplots()```
    - Biến fig đại diện cho toàn bộ hình
    - Biến ax đại diện cho một đồ thị duy nhất trong hình

- Hàm ```plot()```: Vẽ biểu đồ dữ liệu được cung cấp

- Hàm ```show()```: Hiển thị biểu đồ 

````python
import matplotlib.pyplot as plt 

squares = [1,4,9,16,25] 

# Khai báo một đối tượng subplot 
# fig - figure 
# ax - axes
fig, ax = plt.subplots()

ax.plot(squares)

# Đặt tên cho trục x,y và tên biểu đồ
ax.set_xlabel('x')
ax.set_ylabel('y') 
ax.set_title('Biểu đồ')

# Hiển thị biểu đồ 
plt.show()
````

**Changing the Label Type and Line Thickness**

````python
import matplotlib.pyplot as plt 

squares = [1,4,9,16,25] 

# Khai báo một đối tượng subplot 
fig, ax = plt.subplots()

ax.plot(squares,  linewidth = 3)

# Đặt tên cho trục x,y và tên biểu đồ
ax.set_xlabel('x', fontsize = 24)
ax.set_ylabel('y', fontsize = 24 ) 
ax.set_title('Biểu đồ', fontsize = 24)

# Đặt kích thước của nhãn đánh dấu
ax.tick_params(axis = 'both', labelsize = 14) 

# Hiển thị biểu đồ 
plt.show()
````

**Correcting the Plot**

````python
import matplotlib.pyplot as plt 

input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]

fig, ax = plt.subplots() 

ax.plot(input_values, squares) 

ax.set_title('Biểu đồ')
ax.set_xlabel('input_values') 
ax.set_ylabel('squares')
plt.show()
````

**Calculating Data Automatically**

````python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-v0_8')

x = range(1,1001)
y = [i**2 for i in x] 

fig, ax = plt.subplots() 

# Biển diễn mội chuỗi các điểm 
ax.scatter(x,y,s=1)

# Đặt phạm vi cho mỗi trục 
ax.axis([0, 1100, 0, 1100000])

plt.show()
````

**Defining Custom Colors**

Để thay đổi màu sắc của điểm, đối số c trong hàm ```scatter()``` với tên màu được sử dụng trong dấu ngoặc kép

Hoặc 

Sử dụng bảng màu RGB

````python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-v0_8')

x = range(1,1001)
y = [i**2 for i in x] 

fig, ax = plt.subplots() 

# Biển diễn mội chuỗi các điểm 
ax.scatter(x,y,c='r',s=1)

# Đặt phạm vi cho mỗi trục 
ax.axis([0, 1100, 0, 1100000])

plt.show()
````

**Using a Colormap**

````python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-v0_8')

x = range(1,1001)
y = [i**2 for i in x] 

fig, ax = plt.subplots() 

# Biển diễn mội chuỗi các điểm 
ax.scatter(x, y, c=y, cmap=plt.cm.Blues, s=1)

# Đặt phạm vi cho mỗi trục 
ax.axis([0, 1100, 0, 1100000])

plt.show()
````

**Saving Your Plots Automatically**

````python
import matplotlib.pyplot as plt 

plt.style.use('seaborn-v0_8')

x = range(1,1001)
y = [i**2 for i in x] 

fig, ax = plt.subplots() 

# Biển diễn mội chuỗi các điểm 
ax.scatter(x, y, c=y, cmap=plt.cm.Blues, s=1)

# Đặt phạm vi cho mỗi trục 
ax.axis([0, 1100, 0, 1100000])
plt.savefig('squares_plot.png', bbox_inches='tight')
plt.show()
````
