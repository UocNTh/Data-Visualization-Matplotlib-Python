**Scatter Plots**

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='skitscat Raggedy', color='k', s=25, marker="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

**Stack Plots**

Biểu đồ ngăn xếp về cơ bản giống như biểu đồ hình tròn, chỉ theo thời gian.

Hãy xem xét một tình huống mà chúng ta có 24 giờ trong một ngày và chúng ta muốn xem mình đang sử dụng thời gian như thế nào. Chúng ta sẽ chia các hoạt động của mình thành: Ngủ, ăn, làm việc và chơi.

Chúng tôi sẽ giả định rằng chúng tôi đang theo dõi điều này trong vòng 5 ngày, vì vậy dữ liệu ban đầu của chúng tôi sẽ giống như

```python

import matplotlib.pyplot as plt 
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]


plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```
