**The bar chart**

Biểu đồ thanh là một sơ đồ trong đó các biến được biểu diễn dưới dạng các thanh hình chữ nhật, thanh càng cao hoặc càng dài thì giá trị đó càng cao. 

Thông thường, một trục của biểu đồ thanh biểu thị một danh mục và trục còn lại là giá trị của nó. 

Biểu đồ thanh được sử dụng để so sánh dữ liệu rời rạc, chẳng hạn như số lần xuất hiện hoặc tỷ lệ


**Tạo một biểu đồ thanh**


```plt.bar(x, height, width, bottom, align)```


```bottom```: điểm bắt đầu của thanh (mặc định là 0) 

```align```: tên danh mục (mặc định được đặt ở trung tâm)


````python
import matplotlib.pyplot as plt 
plt.figure(figsize = (9,5))
films = ['Tanhaji', 'Baaghi 3', 'Love Aaj Kal', 'Jawaani Jaaneman', 'Street Dancer 3D']  
box_office = [367.65, 137.05, 44.77, 52.63, 97] 
plt.bar(films, box_office)        
plt.show() 
````

**Labels**

````python
plt.figure(figsize = (9,5))
plt.bar(films, box_office)

plt.ylabel('Box Office (crore . ₹)')  # labling y-axis
plt.xlabel('Film title')           # labling x-axis
plt.title('Box office of 5 different films of 2020 in the India')
plt.show()
````


**Gridlines**

Để thêm đường lưới, hãy gọi phương thức grid() và chuyển nó làm tham số cho màu sắc, kiểu đường kẻ, chiều rộng và trục. 


[Xem phim](https://www.geeksforgeeks.org/numpy-linspace-python/)


````python
plt.figure(figsize = (9,5))
plt.bar(films, box_office)

# add grid lines
plt.grid(color='green', linestyle=':', linewidth=1.0, axis='y', alpha=0.5)

plt.ylabel('Box Office (crore . ₹)')  # labling y-axis
plt.xlabel('Film title')           # labling x-axis
plt.title('Box office of 5 different films of 2020 in the India')

plt.show()
````

**Horizontal bar chart** (Biểu đồ thanh ngang)

````python
plt.figure(figsize = (9,5))
plt.barh(films, box_office)

plt.xlabel('Box Office (crore . ₹)')
plt.ylabel('Film title')
plt.title('Box office of 5 different films of 2020 in the India')

plt.show()
````

**Grouped bar plot**

Vẽ nhiều thanh cạnh nhau có thể hữu ích khi chúng ta cần so sánh hai hoặc nhiều chuỗi dữ liệu có chung danh mục.


````python
import matplotlib.pyplot as plt 
import numpy as np 

korea_scores=(554,536,538)
canada_scores=(518,523,525)
china_scores=(413,570,580)
franch_scores=(495,505,499)

index=np.arange(3) # [0,1,2]
bar_width=.2

plt.figure(figsize=(10, 6))

k1=plt.bar(index,korea_scores,bar_width,alpha=.9,label="Korea")
c1=plt.bar(index+bar_width,canada_scores,bar_width,alpha=.9, label="Canada")
ch1=plt.bar(index+bar_width*2,china_scores,bar_width,alpha=.9,label="China")
f1=plt.bar(index+bar_width*3,franch_scores,bar_width,alpha=.9,label="Franch")

# xtinks: nhận và đặt vị trí đánh dấu hiện tại và nhãn của trục x 
# xtinks: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html
plt.xticks(index+.6/2,('Mathematics','Reading','Science'))
plt.ylabel('Mean score in PISA in 2012',fontsize=14)
plt.xlabel('Subjects',fontsize=14)
plt.title('Test scores by Country',fontsize=20)
plt.grid(True)
plt.legend()
plt.show()
````

**Stacked bar plot**


Biểu đồ thanh xếp chồng lên nhau là một loại biểu đồ hiển thị nhiều điểm dữ liệu chồng lên nhau. 

Trong biểu đồ thanh xếp chồng, mỗi thanh đại diện cho một danh mục chứa các danh mục nhỏ hơn. 

Sử dụng để chứng minh cách các bộ phận liên quan đến tổng số tiền.

Trong các ô thanh xếp chồng lên nhau cần cung cấp tham số dưới cùng bổ sung cho biết vị trí bắt đầu của thanh.

````python
years = ['2016', '2017', '2018', '2019']
rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]

plt.figure(figsize=(10, 6))

plt.bar(years, rice, label='Rice')
plt.bar(years, wheat, bottom=rice, label='Wheat')

plt.xlabel('Years', fontsize=14)
plt.ylabel('Preference (%)', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)
plt.legend()
plt.show()
````
